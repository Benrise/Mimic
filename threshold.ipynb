```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
# файл с ботами

path = "./datasets/messages_database_bootcamp_31_01.csv"
df = pd.read_csv(path)
df
```

```python
# файл с ботами со склейкой текста по dialog_id и участнику
df_agg = df[['dialog_id', 'bot_name', 'is_bot', 'text']].groupby(['dialog_id', 'bot_name', 'is_bot']).text.apply(list).reset_index()
df_agg
```

```python
# выгрузить файл со склейкой
df_agg.to_csv("./datasets/bootcamp_agg2.csv")
```

```python
# выгрузить файл без склеек
df[['dialog_id', 'bot_name', 'is_bot', 'text']].to_csv("./datasets/bootcamp2.csv")
```

```python
# файлик с классификаторами
df_class = pd.read_csv("./datasets/classify_messages_database_bootcamp_31_01.csv")

# оставляем своего классификатора
df_class = df_class[df_class.classifier_name=='ArtemUlyana']

# джойнимся к файлу с ботами чтобы получить инфу по итогу бот или не бот
df_full = df_class.merge(df[['dialog_id', 'created_at', 'is_bot']], how = 'left', on = ['dialog_id', 'created_at'])
df_full
```

```python
# задаем порог вероятности, выше которой у нас будет определяться как бот

threshold = 0.5
df_full['final_prob'] = df_full['is_bot_probability'].apply(lambda x: 1 if x > threshold else 0)
accuracy = accuracy_score(df_full['is_bot'], df_full['final_prob'])
accuracy
```

```python
# поиск оптимального порога

def find_optimal_threshold(df, is_bot_probability, is_bot):
    best_threshold = 0
    best_accuracy = 0
    
    # перебираем возможные пороговые значения от 0 до 1 с шагом 0.01
    for threshold in np.arange(0, 1.01, 0.01):
        df['final_prob'] = df['is_bot_probability'].apply(lambda x: 1 if x > threshold else 0)
        accuracy = accuracy_score(df[is_bot], df['final_prob'])
        
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_threshold = threshold
    
    return best_threshold, best_accuracy


optimal_threshold, optimal_accuracy = find_optimal_threshold(df_full, 'is_bot_probability', 'is_bot')

print(f"оптимальный порог: {optimal_threshold:.2f}")
print(f"accuracy: {optimal_accuracy:.2f}")
```
